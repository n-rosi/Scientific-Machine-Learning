```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def reaction_diffusion_rhs(uvt, t, K22, d1, d2, beta, n, N):
    """
    Implements the right-hand side of the reaction-diffusion system.

    Args:
        uvt: A 1D array containing the flattened u and v values.
        t: Time.
        K22: A 1D array containing the squared wavenumbers.
        d1: Diffusion coefficient for u.
        d2: Diffusion coefficient for v.
        beta: Parameter for the omega function.
        n: Spatial grid size.
        N: Total number of grid points.

    Returns:
        A 1D array containing the derivatives of u and v.
    """

    u = uvt[:N].reshape(n, n)
    v = uvt[N:].reshape(n, n)

    A2 = u**2 + v**2
    lam = 1 - A2
    ome = -beta * A2

    # Diffusion terms
    print(u.shape)
    print(K22.shape)
    du_dx2 = np.real(np.fft.ifft2(K22 * np.fft.fft2(u)))
    dv_dx2 = np.real(np.fft.ifft2(K22 * np.fft.fft2(v)))

    # Reaction terms
    du_dt = lam * u - ome * v + d1 * du_dx2
    dv_dt = ome * u + lam * v + d2 * dv_dx2

    return np.concatenate((du_dt.flatten(), dv_dt.flatten()))

t = np.arange(0, 10.05, 0.05)
d1 = 0.1
d2 = 0.1
beta = 1.0
L = 20
n = 512
N = n * n

x = np.linspace(-L/2, L/2, n)
y = x
kx = (2*np.pi/L) * np.fft.fftfreq(n, d=L/n)
ky = kx

X, Y = np.meshgrid(x, y)
KX, KY = np.meshgrid(kx, ky)
K2 = KX**2 + KY**2
K22 = K2#.flatten()

m = 1

u = np.zeros((n, n, len(t)))
v = np.zeros((n, n, len(t)))

r = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)
u[:, :, 0] = np.tanh(r) * np.cos(m * theta - r)
v[:, :, 0] = np.tanh(r) * np.sin(m * theta - r)

uvt0 = np.concatenate((u[:, :, 0].flatten(), v[:, :, 0].flatten()))
uv_sol = odeint(reaction_diffusion_rhs, uvt0, t, args=(K22, d1, d2, beta, n, N))

for j in range(len(t) - 1):
    ut = uv_sol[j, :N].reshape(n, n)
    vt = uv_sol[j, N:].reshape(n, n)
    u[:, :, j + 1] = np.real(np.fft.ifft2(ut))
    v[:, :, j + 1] = np.real(np.fft.ifft2(vt))

    plt.pcolor(x, y, v[:, :, j + 1], cmap='hot')
    plt.colorbar()
    plt.show()

# Plot final state
plt.pcolor(x, y, u[:, :, -1], cmap='hot')
plt.colorbar()
plt.show()
```
